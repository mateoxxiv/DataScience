```python
import numpy as np
import matplotlib.pyplot as plt
```

# Forma

$$f(x)=mx+b \Rightarrow (m,b) \in R$$

**m es igual a**

$$m=\frac{y_2-y_1}{x_2-x_1}$$

**b es:** El punto de corte de la función con el eje y 

# Dominio

$$DOM_f = (-\infty, + \infty)$$

# Código

```python
N = 100
def f(x, m=1, b = 0):
    return m*x+b;
```

```python

x = np.linspace(-10,10,N)

plt.plot(x,f(x, m=-1, b = 5))
plt.axvline(0, c='#000')
plt.axhline(0, c='#000')

plt.grid()
plt.show()
```
