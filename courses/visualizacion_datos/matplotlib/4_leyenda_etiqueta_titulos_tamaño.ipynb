```python
# importar la libreria
import matplotlib.pyplot as plt
import numpy as np
```

```python
# Genermos datos
x = np.linspace(0,10,20)
y = x **2
```

# Generaci√≥n de grafica

set solo se usa para los subsets, para una grafica independiente el comando lo excluye

```python
fig, axes = plt.subplots(1,2, figsize=(8,3))

axes[0].plot(x,y, label='x vs y')
axes[0].set_title('x vs y')
axes[0].set_xlabel('X')
axes[0].set_ylabel('Y')
axes[0].legend()

axes[1].plot(y,x, label='y vs x')
axes[1].set_title('y vs x')
axes[1].set_xlabel('Y')
axes[1].set_ylabel('X')
axes[1].legend()

plt.show()
```
