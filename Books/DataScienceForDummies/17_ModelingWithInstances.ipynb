# Instance modeling

## **Characteristics:**

- Are leazy learners (don´t have a training phase, just remember the data)

- They classify new observation based on a similarity in the training set.

## **Models:**

- k nearest neighbors

- Self organizing maps

- Locally weighted learning

## **Limitations:**


Models of this kind are not suited into data with:

- Noisy data (unexplainable data variation)

- Irrelevant features

- Missing values

## **Nearest Neighbor analysis**

### How it works?

- It takes an observation’s attribute value and then locating another observation whose attribute value is numerically nearest to it.



## **Average Nearest Neighbors**

- The purpose of using an average nearest neighbor algorithm is to classify observations based on the average of the arithmetic distances between them.

- If our goal is to identify and group observations by average similarity, nearest neighbors is a good option

## **K Nearest neighbors**

- Is a supervised machine learning classifier that uses the observations it memorizes from within a test dataset to predict classifications for new, unlabeled observations.

- kNN makes its predictions based on similarity — the more similar the training observations are to the new, incoming observations, the more likely it is that the classifier will assign them both the same class.

- kNN is particularly useful for multi-label learning 

### It predicts better when:

- Data has low noise.

- No outliers

- We have labaled dataset

- Only have relevant features

- Composed of distinguisable groups

- We dont have a large dataset
