# Detecting Outliers

Many statistical and machine learning models assume that there are not ouyliers in our data, thats why it´s so important to remove that group of data.

## Analysing extrem values

Out liers are values that are singnificantly different than the majority of the data

### Types of outliers

1. **points**: values that are significantly different, compared to the normal range of the values in a feature

2. **Contextual**:  data points that are anomalous only within a specific context.

3. **Collective** : These outliers appear nearby to one another, all having similar values that are anomalous to the majority of values in the feature.
