## Importar libreria

```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
import ssl
import urllib.request

# Desactivar verificaci√≥n de certificados SSL
ssl._create_default_https_context = ssl._create_unverified_context
```

## Importar base de datos

```python
df = sns.load_dataset('penguins')
```

```python
p_df = (
    df
    .dropna()
    .copy()
)
```

## Graficas

```python
sns.scatterplot(
    data=p_df,
    x = 'species',
    y='flipper_length_mm',
    hue='species'
)
```

```python
sns.stripplot(
    data=p_df,
    x = 'species',
    y='flipper_length_mm',
    hue='species',
    size=3
)
```

```python
sns.boxplot(
    data=p_df,
    x = 'flipper_length_mm',
    y='species',
    hue='species',
)

sns.stripplot(
    data=p_df,
    x = 'flipper_length_mm',
    y='species',
    hue='species',
    size=3,)
```

```python
sns.violinplot(
    data=p_df,
    x = 'flipper_length_mm',
    y='species',
    hue='species',
)
```

```python
sns.swarmplot(
    data=p_df,
    x = 'flipper_length_mm',
    y='species',
    hue='species')
```
