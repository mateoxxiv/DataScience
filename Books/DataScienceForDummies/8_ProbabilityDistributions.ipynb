# Probability distributions

  Probability distribution is a mathematical function that describes the likelihood of different outcomes in a random experiment or statistical process.

## Properties

- The probability of a single event is [0, 100]

- Probablity of all events sum 100

## Types of probablity distributions

1. Discrete: Values that can be counted by groupingd

2. Continuous: Values in a range of values

## Popular Distributions

1. Normal distribution : these distributions model phenomena that tend toward some most-likely observation (at the top of the bell in the bell curve); observations at the two extremes are less likely.

2. Binomial distribution:  Model the number of successes that can occur in a certain number of attempts when only two outcomes are possible.

3. Categotical distribution: Represents either non- numeric categorical variables or ordinal variables
