# Histograma

Estadística (La forma en la que muestro los datos):

1. 'count': Muestra el número de observaciones en cada bin.
2. 'frequency': Alias para 'count'.
3. 'density': Normaliza de tal manera que el área del histograma se integre a 1.
4. 'probability': Normaliza de tal manera que la suma de los bins del histograma sea 1.
5. 'percent': Normaliza de tal manera que cada bin se muestre como un porcentaje del total, sumando hasta el 100%.
7. 'proportion': Alias para 'probability'.

Múltiples (Cómo trato múltiples valores):

1. 'layer': (Por defecto) Superpone los histogramas para diferentes categorías uno sobre otro. Esto es útil para ver la distribución de cada categoría superpuesta, pero puede resultar en áreas superpuestas.

2. 'dodge': Muestra los histogramas para diferentes categorías uno al lado del otro, evitando la superposición. Esto ayuda a comparar las distribuciones directamente una al lado de la otra.

3. 'stack': Apila los histogramas para diferentes categorías uno sobre otro, similar a 'layer', pero con un enfoque en mostrar la distribución acumulativa.

4. 'fill': Normaliza los histogramas de manera que la suma de las alturas de los bins para cada categoría sea igual a 1 (o al 100% cuando se muestre como porcentajes). Esto es útil para comparar la proporción de cada categoría dentro de los mismos bins.

```python
sns.set_style(style='darkgrid')
sns.histplot(data=tips, x ='tip', bins=20, hue='sex', stat='probability', multiple='dodge')
plt.show()
```

```python
sns.histplot(data=tips, x ='tip', bins=20, cumulative=True)
plt.show()
```

# Kde Plot

```python
sns.kdeplot(data=tips, x ='tip', hue='sex', cumulative=True)
plt.show()
```
