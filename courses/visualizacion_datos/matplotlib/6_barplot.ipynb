```python
# importar la libreria
import matplotlib.pyplot as plt
import numpy as np

country = ['Colombia', 'Venezuela', 'USA']
poblacion = [500, 777, 890]
```

# Es

Un grafico de mucha relevancia en la ciencia de datos, ya que nos ayuda a visualizar datos categoricos

```python
plt.bar(country, poblacion, width=0.5, color=['r','g','b'], alpha=0.5)
plt.xticks(np.arange(0,3),['COL', 'VNZ', 'USA'], rotation=90)
plt.show()
```

```python
plt.barh(country, poblacion, color=['r','g','b'], alpha=0.5)
plt.yticks(np.arange(0,3),['COL', 'VNZ', 'USA'], rotation=90)
plt.show()
```
