# Learning Styles

## Superviesed Algorithms

- Requiere that input data has labaled features.

- The algorithm learn from known features to successfully predict labels from new data

When to use it:

- We have labaled datasets that are good predictors for future events

## Unsupervised Algorithms

- Requiere unlabeled data to group them based on similar features.

## Reinforcement Learning

- It is based on a mechanic similar to how humans and animals learn. The model is given “rewards” based on how it behaves, and it subsequently learns to maximize the sum of its rewards by adapting the decisions it makes to earn as many rewards as possible.
