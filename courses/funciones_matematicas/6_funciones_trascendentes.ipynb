```python
import numpy as np
import matplotlib.pyplot as plt

N = 100
x = np.linspace(-4,4,N)
```

# Son

Funciones que no pueden ser expresadas con polinomios

## Funciones trigonometricas

```python
fig, axes = plt.subplots(nrows=2, ncols=3)
fig.set_size_inches(10,5)

# Plot
axes[0, 0].plot(x*np.pi, np.sin(x))
axes[0, 1].plot(x*np.pi, np.cos(x))
axes[0, 2].plot(x*np.pi, np.tan(x))
axes[1, 0].plot(x*np.pi, np.arcsin(x))
axes[1, 1].plot(x*np.pi, np.arccos(x))
axes[1, 2].plot(x*np.pi, np.arctan(x))

for ax in axes.flat:
    ax.grid(True)

axes[0, 0].set_title('sin')
axes[0, 1].set_title('cos')
axes[0, 2].set_title('tan')
axes[1, 0].set_title('arcsin')
axes[1, 1].set_title('arccos')
axes[1, 2].set_title('arctan')


fig.tight_layout()

plt.show()
```

## Funciones exponenciales

### Forma

$$f(x)=a^x$$  
Donde a es una constante positiva

### Cdigo

```python
N = 100
x = np.linspace(-4,4,N)

plt.plot(x, np.exp(x))
plt.grid()
plt.show()
```

## Funcin Logaritmica

### Forma

$$log_b(x) = n \iff x = n^n$$

### Cdigo

```python
N = 10000
x = np.linspace(-4,4,N)

plt.plot(x, np.log10(x))
plt.grid()
plt.show()
```
