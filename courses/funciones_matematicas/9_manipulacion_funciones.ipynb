```python
import numpy as np
import matplotlib.pyplot as plt

N = 100
x = np.linspace(-10,10,N)
```

# Trasnformaciones

## Horzontales y verticales

1. $y = f(x) + c$, se despaza c unidades hacia arriba.
2. $y = f(x) - c$, se despaza c unidades hacia abajo
3. $y = f(x - c)$, se despaza c unidades hacia la derecha
4. $y = f(x + c)$, se despaza c unidades hacia la izquierda

```python
fig, axes = plt.subplots(ncols=2,nrows=2)
fig.set_size_inches(10,7)

for ax in axes.flat:
    ax.grid(True)
    ax.set_ylim(-5,100)

axes[0,0].plot(x,(x**2) + 5)
axes[0,0].set_title('Desplazamiento arriba')

axes[0,1].plot(x,(x**2) - 5)
axes[0,1].set_title('Desplazamiento abajo')

axes[1,0].plot(x,((x+5)**2))
axes[1,0].set_title('Desplazamiento izquerda')

axes[1,1].plot(x,((x-5)**2))
axes[1,1].set_title('Desplazamiento derecha')

plt.show()
```

## Alargamiento y comprecin

1. $y = f(x) * c$, alarga la grafica verticalmente en un factor de c
2. $y = f(x) * \frac{1}{c}$, comprime la grafica verticalmente en un factor de c
3. $y = f(x * c)$, alarga la grafica horizontalmente en un factor de c
4. $y = f(x * \frac{1}{c} )$, comprime la grafica horizontalmente en un factor de c

```python
fig, axes = plt.subplots(ncols=2,nrows=2)
fig.set_size_inches(10,7)

for ax in axes.flat:
    ax.grid(True)
    ax.set_ylim(0,100)

axes[0,0].plot(x,(x**2) * 5)
axes[0,0].set_title('Alargamiento vertical')

axes[0,1].plot(x,(x**2) * (1/5))
axes[0,1].set_title('Compresin vertical')

axes[1,0].plot(x,((x*5)**2))
axes[1,0].set_title('Alargamiento horizontal')

axes[1,1].plot(x,((x*1/5)**2))
axes[1,1].set_title('Compresin horizontal')

plt.show()
```

## Reflexiones

1. $y =  - f(x)$, refleja respecto a x
2. $y = f(-x)$, refleja respecto a y

```python
fig, axes = plt.subplots(ncols=2,nrows=1)
fig.set_size_inches(10,7)

for ax in axes.flat:
    ax.grid(True)
    ax.set_ylim(-100,100)

axes[0].plot(x, - (x**3))
axes[0].set_title('Reflejo en el eje x')

axes[1].plot(x,((-x)**3))
axes[1].set_title('Reflejo en el eje y')

plt.show()
```
