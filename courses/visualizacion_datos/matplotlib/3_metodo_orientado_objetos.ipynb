```python
# importar la libreria
import matplotlib.pyplot as plt
import numpy as np
```

```python
# Genermos datos
x = np.linspace(0,10,20)
y = x **2
```

# Crear una grafica

1. **figure:** Lienzo en donde vamos a mostrar nuestras graficas
2. **axes:** Graficas que se van a mostrar en el axes ([posicion_x, posicion_y,largo,ancho])

```python
fig = plt.figure()
axes = fig.add_axes([0.1,0.1,0.6,0.5])
axes2 = fig.add_axes([0.2,0.4,0.2,0.2])

axes.plot(x,y)
axes2.plot(y,x)

plt.show()
```

# Subplots

```python
fig, axes = plt.subplots(nrows=1, ncols=2)

# Plot
axes[0].plot(x,y)
axes[1].plot(y,x)
# Grid
axes[0].grid(True)
axes[1].grid(True)

fig.tight_layout()
```
