```python
# importar la libreria
import matplotlib.pyplot as plt
import numpy as np
```

```python
# Genermos datos
x = np.linspace(0,10,20)
y = x **2
```

# Grafico de linea simple

```python
plt.plot(x, y, color='red', marker='+')
plt.grid()
plt.show()
```

# Histograma

```python
plt.hist(x)
plt.grid()
plt.show()
```

# Pie

```python
plt.pie(x)
plt.grid()
plt.show()
```

# Dispersión

```python
plt.scatter(x, y)
plt.grid()
plt.show()
```

# Boxplot

```python
plt.boxplot(x)
plt.grid()
plt.show()
```
