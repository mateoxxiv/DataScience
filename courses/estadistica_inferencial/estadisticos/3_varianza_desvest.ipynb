## Importar las librerias

```python
import pandas as pd
import numpy as no
import matplotlib.pyplot as plt
import seaborn as sns

import ssl
import urllib.request

# Desactivar verificaci√≥n de certificados SSL
ssl._create_default_https_context = ssl._create_unverified_context
```

## Leer la base de datos

```python
url = 'https://archive.ics.uci.edu/ml/machine-learning-databases/iris/iris.data'
df = pd.read_csv(url, header=None, names=['sepal_length', 'sepal_width', 'petal_length', 'petal_width', 'class'])
```

```python
df.head()
```

```python
plt.hist(df['sepal_length'], bins= 20)
plt.show()
```

```python
sns.displot(df, x='sepal_length', kind='hist', bins=20, kde=True)
plt.show()
```

## Calcular varianza

```python
df.sepal_length.var()
```

## Calcular desviaci√≥n estandar

```python
df.sepal_length.std()
```

## Calcular la media

```python
df.sepal_length.mean()
```

## Creamos una muestra aleatoria

```python
muestra = df.sample(frac=0.5)
```

### Calculo de la varianza
