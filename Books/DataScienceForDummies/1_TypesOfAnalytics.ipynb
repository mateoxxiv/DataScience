# Types of analytic


Listed here, in order of increasing complexity, are the four types of data analytics you’ll most likely encounter:

## Descriptive analytics

- It answers "What heppened?"

- It´s based on historical data and current data 

## Diagnostic Analytics

- It answers "why this particular thing happened?" or "what went wrong?"

## Predictive analytics:

- Involve complex model building and analysis in order to predicto future events or trends.

## Perspective analytics:
