```python
# importar la libreria
import matplotlib.pyplot as plt
import numpy as np
```

```python
# Genermos datos
x = np.linspace(0,10,20)
y = x **2
```

# Subplots

```python
# (filas, columnas, indice)

plt.subplot(1,2,1)
plt.plot(x,y,'r*--')
plt.subplot(1,2,2)
plt.plot(x,y,'b+--')
plt.grid()
plt.show()
```
