# Función compuesta

Una función compuesta se obtiene cuando se pasa como argumento de la función principal el valor de otra función, en este ejemplo tenemos

## Forma

$$ f \circ g = (f \circ g)(x) = f(g(x)) $$

## Código

```python
def g(x):
    return x**2

def f(X):
    return np.cos(x)

y = g(x)

f_o_g = f(g(x))

plt.plot(x, f_o_g)
plt.grid()
plt.show()
```
