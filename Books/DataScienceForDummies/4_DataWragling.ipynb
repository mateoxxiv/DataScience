# Data Wragling

Is the process and procedures that are use to clean and convert data fron its raw format and structure to another, in order to get accurate data form for tools and scripts

## Data extraction - ( data mining )

- Identify wich dataset are relevant to the problem

- Extract the rigth quantity of data to solve the problem

## Data preparation

1. **Import**: read relevant datasets into our application

2. **Clean**: Remove strays data, duplicates and out-of.range records. | Standarizing case.

3. **Transform**: Missing values, deal with out-liers and scale variables

4. **Process**: Parsing, recoding, concatenation and other methods to reformat our data

5. **Log in**: Descripcion of our data set, like descriptive statistics, variables format and information, collection methods and more.

6. **Back up**: Save a backup of the clean version of the data set
