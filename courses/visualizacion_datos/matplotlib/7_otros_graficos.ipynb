```python
# importar la libreria
import matplotlib.pyplot as plt
import numpy as np

x = np.linspace(1,10,100)
data = np.random.randint(1,20,200)
data2 = np.random.randint(1,20,200)

print(plt.style.available)
```

# Histograma

```python
plt.style.use('Solarize_Light2')

plt.figure(figsize=(6,3))
plt.hist(data, bins=10, alpha=0.6, histtype='barstacked')
plt.show()
```

# Boxplot

```python
plt.figure(figsize=(6,3))
plt.boxplot(data)
plt.show()
```

```python
plt.figure(figsize=(6,3))
plt.boxplot(data, vert=False, patch_artist=True, notch=True, showfliers=True)
plt.show()
```

# Scatter

```python
plt.scatter(data,data2, s=(data+data2)**1.5, c=data)
plt.show()
```
