# Regressions

They are usefull in these cases:

1. Detemrine the strength of the correlation between variables.

2. Predict future values from historical data (Predicting future values from historical ones will generate incorrect results when present circumstances change.)

## Linear Regression

Linear regression is a machine learning method you can use to describe and quantify the relationship between your target variable, y — the predictant, in statistics lingo — and the dataset features you’ve chosen to use as predictor variables (commonly designated as dataset X in machine learning).

### Considerations

1. Only works with numerical variables

2. Missing values cause problems

3. Outliers produce inaccurancy

4. Asusmes that features have linear relatoinship between features and target value. (in case that not we can use log transformation)

5. All features are independent for each other

6. Predicted errors should be normally distributed

## Logistic Regression

- Is a machine learning method you can use to estimate values for a categorical target variable based on your selected features. Your target variable should be numeric, and contain values that describe the target’s class — or category.

- In addition to predicting the class of observations in your target variable, it indicates the probability for each of its estimates

## Ordinary least squares (OLS)

It´s a method that fit a linear regression line to a dataset.

It works by squaring the vertical distance that describe the distances between the data points and the best fit lines, minimizing the sum of the squared distace
