## Identifying clusters in our data

### **k-means**

#### Characteristics: 

- Makes its predictions based on the number of the number of centroids present (k) and the nearest mean values (measure by euclidean distance)

- Our feateres must be in the same scale of values, so sacle the values before predicting useing k-means.

#### How to choose the k number of clusters

- If our data set has more or less 4 features, we can determiniate our k values looking at a scatter plot

- **Silhouette coefficient**: alculates the average distance of each point from all other points in a cluster and then compares that value with the average distance to every point in every other cluster

#### How it works:

1. **Initialization of Cluster Centers**:

    - Initially place sample cluster centers on an n-dimensional plot. This can be done manually or randomly.

    - The placement of these centers is crucial as it sets the starting points for the algorithm.


2. **Evaluation of Center Movement**:

    - Assess whether moving each cluster center in any single direction would result in a new center with higher density.

    - The goal is to move the centers from regions of lower density to regions of higher density, optimizing the clustering.


3. **Iterative Process**:

    - Repeat the process of evaluating and moving cluster centers iteratively.

    - The centers are adjusted to maximize the density of observations around each center until a local maximum density is achieved.


4. **Convergence to True Cluster Centers**:

    - Continue the iterative process until all cluster centers are within a region of local maximum density.

    - At this point, each cluster ideally has a maximum number of points closest to its cluster center, representing a more optimal solution.


### **Kernel density estimation - KDE**

#### How it works

1. **Función de Suavizado de Densidad:**
   - KDE es un método de suavizado de densidad que cuantifica la densidad de los puntos de datos al colocar un núcleo (una función de ponderación) en cada punto de datos.
   - La idea es generar una estimación suave de la función de densidad de probabilidad para todo el conjunto de datos.

2. **Colocación y Suma de Núcleos:**
   - Para cada punto de datos en el conjunto de datos, se coloca un núcleo y luego se suman estos núcleos para crear una estimación de densidad para toda la región.
   - La elección de la función del núcleo y sus parámetros (ancho de banda) es crucial y puede afectar la sensibilidad del método.

3. **Representación Espacial de la Densidad:**
   - KDE proporciona una representación espacial de la densidad de puntos de datos en todo el conjunto de datos.
   - Los valores de densidad de núcleo más altos indican regiones de mayor densidad de puntos, mientras que los valores más bajos indican regiones con menos puntos de datos.

4. **Identificación de Clústeres:**
   - Los clústeres en los datos se pueden identificar localizando áreas con una alta densidad de núcleo, que representan regiones donde los puntos de datos están densamente concentrados.
   - A diferencia de K-means, KDE no requiere la suposición de un número fijo de clústeres y es capaz de capturar formas y estructuras complejas en los datos.


### **Hierarchical algorithms**

- A hierarchical clustering algorithm is yet another alternative to k-means clustering

- In comparison to k-means clustering, the hierarchical clustering algorithm is a slower, clunkier unsupervised clustering algorithm.

- It predicts groupings within a dataset by calculating the distance and generating a link between each singular observation and its nearest neighbor.
