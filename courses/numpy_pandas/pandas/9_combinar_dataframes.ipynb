```python
import pandas as pd
df = pd.read_csv('./usuarios.csv')
df2 =  pd.read_csv('./usuarios.csv')
df
```

# Concatenar

```python
# Por defecto me concatena por linea
pd.concat([df,df2], ignore_index=True)
```

```python
# Concatenar columnas
pd.concat([df,df2], ignore_index=True, axis=1)
```

# Combinar

```python
df.merge(df2, how='inner' , left_on='id', right_on='id', suffixes=['_left','_rigth'])
```

# Combinar con indices

```python
df.join(df2, lsuffix='_df1', rsuffix='_df2')
```
