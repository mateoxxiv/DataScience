## Importar libreria

```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
import statsmodels.formula.api as smf
import ssl
import urllib.request

# Desactivar verificaci√≥n de certificados SSL
ssl._create_default_https_context = ssl._create_unverified_context
```

## Importar base de datos

```python
df = sns.load_dataset('penguins')
```

```python
p_df = (
    df
    .dropna()
    .copy()
)

p_df['number_sex'] = p_df['sex'].map({'Male':1, 'Female':0})

p_df.head()
```

## Modelo 1

```python
modelo_1 = (smf.ols(
    formula='body_mass_g ~ bill_length_mm + bill_depth_mm',
    data=p_df
).fit())

modelo_1.summary()
```

```python
modelo_1 = (smf.ols(
    formula='body_mass_g ~ bill_length_mm + bill_depth_mm + flipper_length_mm + number_sex',
    data=p_df
).fit())

modelo_1.summary()
```

```python
modelo_2 =  (
    smf.ols(
        formula='body_mass_g ~ flipper_length_mm',
        data=p_df
    ).fit()
)
modelo_2.summary()
```
