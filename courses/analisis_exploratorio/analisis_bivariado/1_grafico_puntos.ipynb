## Importar libreria

```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
import ssl
import urllib.request

# Desactivar verificaci√≥n de certificados SSL
ssl._create_default_https_context = ssl._create_unverified_context
```

## Importar base de datos

```python
df = sns.load_dataset('penguins')
```

```python
p_df = (
    df
    .dropna()
    .copy()
)
```

## Recomendaicones para contruir un mejor grafico de puntos

1. Modificar la transparencia de los puntos
2. A travez de histograma

```python
sns.scatterplot(
    data=p_df,
    x='bill_length_mm',
    y='bill_depth_mm',
    alpha=1/2,
    hue='species'
)
```

```python
sns.displot(
    data=p_df,
    x='bill_length_mm',
    y='bill_depth_mm',
    hue='species',
    rug=True
)
```

```python
sns.displot(
    data=p_df,
    x='bill_length_mm',
    y='bill_depth_mm',
    hue='species',
    rug=True,
    kind='kde'
)
```

```python
sns.jointplot(
    data=p_df,
    x='bill_length_mm',
    y='bill_depth_mm',
    hue='species'
)
```
