### Map reducing

Works by converting data into sets of tuples, using parallel distribued computing to transform big data into manageable data, the process  works as following:

#### Map the data

- Incoming data is first transformed into key-value pairs and divided into fragments.

- Fragments are assigned to map tasks within computing clusters.

- Each cluster is allocated a number of map tasks, which are distributed among its nodes.

- Processing generates intermediate key-value pairs.

- Intermediate pairs are sorted by key values and divided into new fragments.

- Count of these new fragments equals the count of reduce tasks.


### Reduce the data

- Each reduce task is assigned a fragment.

- Reduce tasks process their assigned fragment, producing output key-value pairs.

- Tasks are distributed across cluster nodes.

- Upon completion, final output is written onto a file system.
