# Popular ML methods in business

## Linear Regression

It´s usefull in cases as:

- Predictions for sales forecasts.

- Pricing optimization.

- Marketing optimization.

- Financial risk assessment.

## Logistic Regression

It´s usefull in cases as:

- Predict customer churn.

- Predict response-versus-ad spending.

- Predict the lifetime value of a customer.

- Monitor how business decisions affect predicted churn rates.

## Naive Bayes

It´s usefull in cases as:

- Spam detector

- Analyze customer sentiment

- Automatically categorize products, customers, or competitors

## K-means Clustering

It´s usefull in cases as:

- Cost modeling

- Customer segmentation

## Hierarchical Clustering

It´s usefull in cases as:

- model business processes

- Segment customers based on survey responses.


## K-nearest neighbor Classication

It´s usefull in cases as:

- text document classification

- Financial distress prediction modeling

- competitor analysis and classification.

## Principal Component Analysis

It´s usefull in cases as:

- Detecting fraud

- Speech recognition

- Spam detection.
