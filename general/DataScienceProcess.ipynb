# Step by step DataScience process

Data science, as the name suggests, is a systematic approach to extracting valuable insights from data. It involves a combination of mathematical techniques, statistical analysis, computer science methodologies, and business acumen. The primary objective of this interdisciplinary field is to uncover opportunities, forecast future events, and facilitate informed decision-making based on robust criteria.

In the realm of scientific inquiry, Data Science adheres to a structured process that guides us from gathering data from diverse sources to assessing and refining specific models deployed in production environments. Below, I will outline this process:

- **Understanding the Business** – Clarify the problem and goals.

- **Defining the Problem** – Specify the precise goals the data science project aims to accomplish.

- **Getting the Data** – Gather relevant data from different sources.

- **Preparing the Data** – Cleaning and formatting for analysis.

- **Building Models** - Applying analytics like machine learning to get insights.

- **Evaluating Models** - Testing to pick the best performers.

- **Deploying the Solution** – Launching it for business use.

### Modeling

When selecting a model for a particular problem in data science, it's crucial to choose the one that best fits the problem at hand. Once a model is chosen, it's essential to provide two reports:

1. **Model Specification Report:**

- Name of the model (e.g., Linear Regression, Random Forest, Neural Network)

- Description of the model architecture or algorithm used

- Parameters and hyperparameters of the model

- Assumptions underlying the model

- For example, if using a Linear Regression model, the report would specify that it's a linear model assuming a linear relationship between 
- the input features and the target variable.



2. **Assumptions Report:**

- Linearity: Assumes that the relationship between predictors and the target variable is linear.

- Independence: Assumes that observations are independent of each other.

- Homoscedasticity: Assumes that the variance of errors is constant across all levels of the predictor variables.

- Normality: Assumes that the residuals of the model are normally distributed.

- No multicollinearity: Assumes that there is no multicollinearity among the predictor variables.

- For instance, if using a Linear Regression model, assumptions might include linearity of relationships between predictors and the target - 
variable, normality of residuals, and absence of multicollinearity among predictors.


The main objective of these step is to train and test the model and avoiding two problems:

1. **Overfitting**: is when a model describe erfectly a data set but when it comes to other data it lack of presicion.

2. **Underfitting**: our model is not predicting anything at all.

Finally we have to optimize our model with the correct parameters.
