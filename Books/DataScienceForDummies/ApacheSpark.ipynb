# Apache Spark

Open source: http://spark.apache.org/downloads.html

## Spark SQL:

- Used for working with and querying structured data in Spark.

- Allows the use of Spark's built-in SQL package or HiveQL for queries if working with structured data in Hive.

## Streaming

- Module for big data processing, breaking continuously streaming data into smaller data streams called Dstreams.

- Performs microbatch processing in three-second cycles for efficient handling of data.

## MLlib

- Submodule for data analysis, statistics generation, and deploying machine learning algorithms within the Spark environment.

- APIs available in Java, Scala, Python, and R.

- Enables building machine learning models directly from HDFS, reducing reliance on data engineers.

- Computations known to be 100 times faster when processed in-memory using Spark compared to the standard MapReduce framework.
