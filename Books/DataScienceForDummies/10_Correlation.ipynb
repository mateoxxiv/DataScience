# Quantifying Correlation

- Many statisticals and machine learning models assume our features are independent.

- We need to test the correlation of our features

## Pearson´s R

It assumes:

1. Data is normally distributed

2. Continuous, numeric data

3. Variables are linearly related

## Spearman´s rank correlation

it assumes:

1. Variables are ordinal

2. Variables related non-linearly

3. Data is not normally distributed]
