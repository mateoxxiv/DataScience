# SVD

Decomposes a Matrix (X) into three resultant matrices:

- X: Original matrix.

- U: left singular matrix, holds the non-redundant information about our observations

- V: rigth singular matrix, holds the non-redundant information about the columns

- S: Squeared root of eigen values, conatains information about the procedures performed during the compression

## Factor analysis

- The variance measure how much the values of our variable variate from our mean.

- The grater the variance, the more information the variable has.

This method asumes:

1. Our features are numeric variables

2. Features are continuous or ordinal

3. We have at least 100 observations in our data base

4. Our data is homogenous

5. There is a correlation r > 0.3between our features.

## Principal component analysis

Finds relationships between features in your dataset and then transforms and reduces them to a set of non-information-redundant principle components — uncorrelated features that embody and explain the information that’s contained within the dataset.
