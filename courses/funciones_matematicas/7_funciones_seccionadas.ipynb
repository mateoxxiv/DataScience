```python
import numpy as np
import matplotlib.pyplot as plt

N = 100
x = np.linspace(-4,4,N)
```

# Funciones seccionadas

Son funciones que se comportan de una forma diferente dependiendo unos intervalos

$$
H(x) =
\begin{cases} 
0 & \text{si } x < 0 \\
1 & \text{si } x \geq 1 \\
\end{cases}
\text{, en este ejemplo se muestra la función escalón de Heaviside}$$


```python
def H(x):
    return [0 if i < 0 else 1 for i in x]
        

plt.plot(x, H(x))
plt.grid()
plt.show()
```
