## Importar libreria

```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
import scipy.stats as stats
import empiricaldist
import ssl
import urllib.request

# Desactivar verificaci√≥n de certificados SSL
ssl._create_default_https_context = ssl._create_unverified_context
```

## Importar base de datos

```python
df = sns.load_dataset('penguins')
```

```python
p_df = (
    df
    .dropna()
    .copy()
)
```

## Funciones de densidad de probabilidad

```python
sns.kdeplot(
    data=p_df,
    x='flipper_length_mm',
    bw_method=0.1)
```

```python
p_stats = p_df.body_mass_g.describe()
```

```python
xs = np.linspace(p_stats['min'], p_stats['max'])
ys = stats.norm(p_stats['mean'], p_stats['std']).cdf(xs)
```

```python
empiricaldist.Cdf.from_seq(p_df.body_mass_g).plot()
plt.plot(xs, ys)
```

```python
xs = np.linspace(p_stats['min'], p_stats['max'])
ys = stats.norm(p_stats['mean'], p_stats['std']).pdf(xs)
plt.plot(xs, ys)
sns.kdeplot(
    data=p_df,
    x='body_mass_g',
    )
```
